# Explanatory Visualization CodeAlong

- Life Expectancy Data: https://www.kaggle.com/kumarajarshi/life-expectancy-who

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

## Csv file
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRQpH02vlgxAwATTUhJBC0MGiciSz-vUPenWNbrkVH4ijb12NXK-4ut0jLqbANnBgRUo36ZAXDfeQKa/pub?output=csv"
```

```python
df = pd.read_csv(url)
display(df.info())
df
```

# Q1: Does education increase life expectancy?


```python
sns.set_context('talk')
```

```python
ax = sns.scatterplot(data=df, x='Schooling',y='Life expectancy', hue='Life expectancy',
                     )
ax.legend().remove()
```

```python
ax = sns.scatterplot(data=df, x='Schooling',y='Life expectancy', hue='Life expectancy')
ax.legend().remove()
```

```python
ax = sns.scatterplot(data=df, x='Schooling',y='Life expectancy', hue='Status')
ax.legend(bbox_to_anchor=[1,1])
```

```python
sns.regplot(data=df, x='Schooling',y='Life expectancy', line_kws=dict(color='black',
                                                                      ls=':'))
```

```python
ax = sns.scatterplot(data=df, x='Schooling',y='Life expectancy', hue='Life expectancy',palette='Reds',
                     )
ax.legend().remove()
ax.set_title("Effect of Schooling on Life Expectancy ");
```

> As Schooling increases, so does Life Expectancy, overall. 

## Q2: Does the development status of a country affect schooling or life expectancy?


```python
df.nunique()
```

```python
sns.barplot(data=df, x='Status', y='Life expectancy')
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(12,4))

sns.barplot(data=df, x='Status', y='Life expectancy',ax=axes[0])

sns.barplot(data=df, x='Status', y='Schooling',ax=axes[1])

fig.suptitle("Comparing Developed vs Developing Countries")
```

```python
fig, axes = plt.subplots(nrows=2, sharex=True, figsize=(5,10))

sns.barplot(data=df, x='Status', y='Life expectancy',ax=axes[0])
axes[0].set_xlabel('')

sns.barplot(data=df, x='Status', y='Schooling',ax=axes[1])

fig.suptitle("Effect of Development Status on\n Schooling or Life Expectancy",y=1.05)
plt.tight_layout()
```

> Developed countries have higher average education levels and life expetancy.

```python
fig, axes = plt.subplots(nrows = 2, figsize = (5,10), sharex = True)
sns.boxplot(data = df, y = 'Life expectancy', x = 'Status', ax = axes[0])
sns.boxplot(data = df, y = 'Schooling', x = 'Status', ax = axes[1])
axes[0].set_xlabel('')
axes[1].set_xlabel('')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 2, figsize = (5,10), sharex = True)
sns.swarmplot(data = df, y = 'Life expectancy', x = 'Status', ax = axes[0])
sns.swarmplot(data = df, y = 'Schooling', x = 'Status', ax = axes[1])
axes[0].set_xlabel('')
axes[1].set_xlabel('')
plt.tight_layout()
```

# Post-Class

- Exploring the groupby approach for the final Year question

```python
year_means = df.groupby("Year").mean()
year_means
```

```python
fig, axes = plt.subplots(nrows=3, sharex=True,figsize=(15,10))
year_means.plot(y='Life expectancy',ax=axes[0],marker='o')
year_means.plot(y='Schooling',ax=axes[1],marker='o')
year_means.plot(y='GDP',ax=axes[2],marker='o')
[ax.grid() for ax in axes];

fig.tight_layout()
```

```python
plt.subplots(1,2)
```

```python
fig, axes = plt.subplots(nrows=2, figsize=(15,10))

axes[0].plot(year_means.index, year_means['Life expectancy'])
axes[0].set_xticks(year_means.index);
axes[0].grid()
```
